```python
##Fill NAs
def impute_missing_value(df):
  df.fillna({'age_cohort':'unknown','marital_status':'unknown'}, inplace=True)
  df.fillna({'home_is_rental':'0','n_family':'1','no_of_children':'0','income_range':df.income_range.mean()}, inplace=True)
  return df 
```

```python
##Function to convert date into datetime format
def Change_to_datetime(date_col):
    new_datetype = []
    for i, date in enumerate(date_col):
        date_format = "%m/%d/%Y" 
    
        date= date[:-2]+'20'+date[-2:]#change the string format to the format convirtable to datetype object
        date= datetime.strptime(date, date_format)
        new_datetype.append(date)
        
    new_datetype = pd.Series(new_datetype)
    return new_datetype
```

```python
## Function to do some date transformation
def dataframe_transform(training):
  end_date =Change_to_datetime(training.end_date)
  training['end_date'] = end_date
  start_date =Change_to_datetime(training.start_date)
  training['start_date'] = start_date
  training['date'] = pd.to_datetime(training['date'])
  func = lambda x: x.days
  training['duration_days']=training.end_date - training.start_date
  training['duration_days'] = training['duration_days'].apply(func)
  training['coupon_used']= training.dscnt_pcode.apply(lambda x:1 if x !=0 else 0)
  return training
```

```python
## Clean dataframe, removing irelevant rows by logic 
def clean_dataframe(training):
  training.sort_values(by='date') #sort df by date before split
  mask = (training['date'] >= training['start_date']) & (training['date'] <= training['end_date'])
  trans_valid_prom= training.loc[mask]
  valid_list = training.loc[mask].index.to_list()
  trans_expire_prom= training[~training.index.isin(valid_list)]
  invalid_index = (trans_expire_prom[trans_expire_prom['redeemed'] == 1]).index.to_list()
  training = training[~training.index.isin(invalid_index)]
  problematic_data = training.query('redeemed ==1 and dscnt_store ==0 and dscnt_pcode ==0')
  problematic_data_index= problematic_data.index.to_list()
  training = training[~training.index.isin(problematic_data_index)]
  return training
```

```python
##Aggregate dataframe by different indexes to reduce the total size
def aggregate_data(training):
  pivot_cust_redeemed = pd.pivot_table(training,index =['prod_id','cust_id','c_id','redeemed'],
                                      values = ['manuf', 'brand_type', 'category',
        'age_cohort', 'marital_status', 'home_is_rental', 'n_family',
        'no_of_children', 'income_range', 'duration_days', 'coupon_used','qty','price','date','dscnt_store','dscnt_pcode','c_type']
                                      ,aggfunc = {
                          'qty':np.mean,
                          'price':np.mean,
                          'dscnt_store':np.mean,
                          'dscnt_pcode':np.mean,
                          'coupon_used': np.sum,
                          'manuf':lambda x: x.value_counts().index[:1],
                          'brand_type':lambda x: x.value_counts().index[:1], 
                          'category':lambda x: x.value_counts().index[:1],
                          'age_cohort':lambda x: x.value_counts().index[:1],
                          'marital_status':lambda x: x.value_counts().index[:1],
                          'home_is_rental':lambda x: x.value_counts().index[:1],
                          'n_family':lambda x: x.value_counts().index[:1],
                          'no_of_children':lambda x: x.value_counts().index[:1],
                          'income_range':lambda x: x.value_counts().index[:1],
                          'duration_days':lambda x: x.value_counts().index[:1],
                          'c_type':lambda x: x.value_counts().index[:1],
                          'date' : lambda x:x.value_counts().index[:1]
                          })
  final_data = pivot_cust_redeemed.reset_index()
  return final_data
```

```python
## Rename the column names
def rename_cols(final_data):
  new_cols = {'age_cohort':'mostfrequent_age_cohort',
              'brand_type':'mostfrequent_brand_type',
              'category':'mostfrequent_category',
              'coupon_used':'total_coupoun_used',
              'dscnt_pcode':'avg_dcnt_pcode',
              'dscnt_store':'avg_dscnt_store',
            'duration_days':'mostfrequent_duration_days',
            'home_is_rental':'mostfrequent_home_is_rental',
            'income_range':'mostfrequent_income_range',
            'manuf':'mostfrequent_manuf',
            'marital_status':'mostfrequent_marital_status',
            'n_family':'mostfrequent_n_family',
              'no_of_children':'mostfrequent_no_of_children',
              'price':'avg_price',
              'qty':'avg_qty',
              'c_type':'mostfrequent_ctype'}
  final_data.rename(columns=new_cols, inplace = True)
  final_data_copy = final_data.copy()
  final_data_copy['mostfrequent_n_family'] = final_data_copy['mostfrequent_n_family'].astype(str)
  final_data_copy['mostfrequent_n_family'] = final_data_copy['mostfrequent_n_family'].replace({'1': '1.0', '2':'2.0', '3':'3.0', '4':'4.0'})
  final_data_copy.date = pd.to_datetime(final_data_copy.date, yearfirst=True)
  final_data_copy['date']=final_data_copy['date'].apply(lambda x: x.year)
  return final_data_copy
```

```python
###Dummy columns
def dummy_(final_data_copy):
  final_data_copy= pd.get_dummies(data=final_data_copy, columns=['mostfrequent_age_cohort']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_brand_type']
        ,drop_first=True,dummy_na=False)
  final_data_copy= pd.get_dummies(data=final_data_copy, columns=['mostfrequent_ctype']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_category']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_home_is_rental']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_marital_status']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_n_family']
        ,drop_first=True,dummy_na=False)
  final_data_copy=pd.get_dummies(data=final_data_copy, columns=['mostfrequent_no_of_children']
        ,drop_first=True,dummy_na=False)
  return final_data_copy
```

```python
###Split the data 
def split_data(final_data_copy):
  x =final_data_copy.loc[:, final_data_copy.columns != 'redeemed']
  x= x.drop(columns=['date'])
  y=final_data_copy['redeemed']
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4,shuffle= False)
  return X_train, X_test, y_train, y_test
```

```python
## Split the test data into test1 and test 2, one for randomized search cv, one for final test
def split_test(X_test,y_test):
  X_test1, X_test2, y_test1, y_test2 = train_test_split(X_test, y_test, test_size=0.5,shuffle= False)
  return X_test1, X_test2, y_test1, y_test2
```

```python
## Smote to balance the dataset
def smote_(X_train, X_test, y_train, y_test):
# Now let's compare a few different ratios of minority class to majority class
  ratios = [0.15, 0.25, 0.33, 0.5, 0.7, 1]
  names = ['0.15', '0.25', '0.33','0.5','0.7','even']
  colors = sns.color_palette('Set2')

  plt.figure(figsize=(10, 8))

  for n, ratio in enumerate(ratios):
      # Fit a model
      smote = SMOTE(sampling_strategy=ratio)
      X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
      print(X_train_resampled)
      logreg = LogisticRegression(fit_intercept=False, C=1e20, solver ='lbfgs')
      model_log = logreg.fit(X_train_resampled, y_train_resampled)
      print(model_log)

      # Predict
      y_hat_test = logreg.predict(X_test)

      y_score = logreg.decision_function(X_test)

      fpr, tpr, thresholds = roc_curve(y_test, y_score)
      
      print('AUC for {}: {}'.format(names[n], auc(fpr, tpr)))
      print('-------------------------------------------------------------------------------------')
      lw = 2
      plt.plot(fpr, tpr, color=colors[n],
              lw=lw, label='ROC curve {}, AUC:{}'.format(names[n], round(auc(fpr, tpr),3)))

  plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])

  plt.yticks([i/20.0 for i in range(21)])
  plt.xticks([i/20.0 for i in range(21)])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.title('Receiver operating characteristic (ROC) Curve')
  plt.legend(loc='lower right')
  plt.show()
  return X_train_resampled, y_train_resampled
```

```python
##Standardlize the values to reduce the error
def Standardlize_(X_train_resampled,X_test1):
  scaler = StandardScaler()
  scaled_X_train_resampled = scaler.fit_transform(X_train_resampled)
  scaled_X_test = scaler.transform(X_test1)
  scaled_X_train_resampled = pd.DataFrame(scaled_X_train_resampled, columns = X_train_resampled.columns)
  scaled_X_test = pd.DataFrame(scaled_X_test, columns = X_train_resampled.columns)
  return scaled_X_train_resampled, scaled_X_test
```

```python
def grid_search(scaled_X_train_resampled, y_train_resampled,scaled_X_test,y_test):
  pipe_dt = Pipeline([('DT',DecisionTreeClassifier())])
  pipe_lr = Pipeline([('LR', LogisticRegression(solver='lbfgs', max_iter=1000))])
  pipe_svm = Pipeline([('SVM', svm.SVC())])
  pipe_xgb = Pipeline([('xgboost',XGBClassifier())])
  pipe_GaussianNB = Pipeline([('gussiannb',GaussianNB())])
  pipe_randomforest = Pipeline([('randforesr',RandomForestClassifier())])

  param_range = [1, 2, 3, 4, 5, 6]
  param_range_fl = [1.0, 0.5, 0.1]
  n_estimators = [50,100,150]
  learning_rates = [.1,.2,.3]
  lr_param_grid = [{
                    'LR__C': param_range_fl}]
  dt_param_grid = [{'DT__min_samples_leaf': [1,2,4,8,20]}]
  svm_param_grid = [{'SVM__kernel': ['linear', 'rbf'], 
                      'SVM__C': param_range}]
  xgb_boost_grid = { 'xgboost__max_depth': [3, 5, 6, 10, 15, 20],
            'xgboost__learning_rate': [0.01, 0.1],
            'xgboost__subsample': np.arange(0.5, 1.0, 0.1),
            'xgboost__colsample_bytree': np.arange(0.4, 1.0, 0.2),
            'xgboost__colsample_bylevel': np.arange(0.4, 1.0, 0.2),
            'xgboost__n_estimators': [100,300, 500],
            'xgboost__scale_pos_weight':[1,2,3,4,5,6,7,8]}
          

  random_forest_grid = {
  'randforesr__max_depth': [10, 20, 30, 40],
  'randforesr__max_features': ['auto', 'sqrt'],
  'randforesr__min_samples_leaf': [1, 2, 4],
  'randforesr__min_samples_split': [2, 5, 10],
  'randforesr__n_estimators': [200, 400, 600]}

  GaussianNB_param_grid ={'gussiannb__kernel': ['rbf','linear'],
                          'gussiannb__C':[0.05,0.1,0.2],
                            'gussiannb__gamma':[0.1,0.2,0.3]}
  
  

  lr_grid_search = RandomizedSearchCV(estimator=pipe_lr,
          param_distributions=lr_param_grid,
          scoring='recall')
  dt_grid_search = RandomizedSearchCV(estimator=pipe_dt,
          param_distributions=dt_param_grid,
          scoring='recall')

  svm_grid_search = RandomizedSearchCV(estimator=pipe_svm,
          param_distributions=svm_param_grid,
          scoring='recall')
  xgb_grid_search = RandomizedSearchCV(estimator = pipe_xgb,
                                param_distributions = xgb_boost_grid,
                                scoring='recall')
  rf_grid_search = RandomizedSearchCV(estimator=pipe_randomforest,
                                param_distributions=random_forest_grid,
                                scoring='recall')



  grids = [lr_grid_search, dt_grid_search, svm_grid_search,xgb_grid_search,rf_grid_search]
  for pipe in grids:
      pipe.fit(scaled_X_train_resampled, y_train_resampled)
    
  grid_dict = {0: 'Logistic Regression', 1: 'Decision Trees',  
              2: 'Support Vector Machines',3:'XGBOOST',4:'randomforest'}
  for i, model in enumerate(grids):
      print('{} Recall is: {}'.format(grid_dict[i],
      model.score(scaled_X_test,y_test)))
      print('{} Best Params: {}'.format(grid_dict[i],model.best_params_))
```
